```python
# 1. Function to find maximum of three numbers

def maximum(a, b, c):
    return max(a, b, c)

print("Maximum:", maximum(10, 25, 15))
```

```python
# 2. Function to return distinct elements from a list

def distinct_list(lst):
    return list(set(lst))

print("Distinct List:", distinct_list([1, 2, 2, 3, 4, 4, 5]))
```

```python
# 3. Function to multiply all numbers in a list

def multiply_list(lst):
    result = 1
    for i in lst:
        result *= i
    return result

print("Multiplication:", multiply_list([2, 3, 4]))
```

```python
# 4. Function to calculate factorial

def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

print("Factorial:", factorial(5))
```

```python
# 5. Program to reverse a string

def reverse_string(s):
    return s[::-1]

print("Reversed String:", reverse_string("Python"))
```

```python
# 6. Function to check whether number falls within range

def check_range(num, start, end):
    if start <= num <= end:
        return "Number is within range"
    else:
        return "Number is outside range"

print(check_range(7, 1, 10))
```

```python
# 7. Function to print even numbers from a list

def even_numbers(lst):
    print("Even Numbers:")
    for i in lst:
        if i % 2 == 0:
            print(i)

even_numbers([1, 2, 3, 4, 5, 6, 7, 8])
```

```python
# 8. Function to check whether number is prime

def is_prime(n):
    if n <= 1:
        return "Not Prime"

    for i in range(2, n):
        if n % i == 0:
            return "Not Prime"

    return "Prime"

print("Prime Check:", is_prime(11))
```

```python
# 9. Function to count upper and lower case letters

def count_case(s):
    upper = 0
    lower = 0

    for ch in s:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1

    print("Upper Case Letters:", upper)
    print("Lower Case Letters:", lower)

count_case("Hello Python")
```
