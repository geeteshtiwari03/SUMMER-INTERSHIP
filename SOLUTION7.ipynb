```python
import pandas as pd
# This code creates a pandas Series from a dictionary and prints it.
data = {
    "a": 10,
    "b": 20,
    "c": 30
}

s1 = pd.Series(data)

print(s1)
```

```python
# This code creates a pandas Series from a list and prints it.
l = [100, 200, 300, 400]

s2 = pd.Series(l)

print(s2)
```

```python
# Accessing elements of the Series
print(s2[0])      # first element
print(s2[2])      # third element
```

```python
# This code creates a pandas DataFrame from a list of lists and prints it.
data = [
    [1, "John", 50000],
    [2, "Sam", 60000],
    [3, "Alice", 70000]
]

df1 = pd.DataFrame(data, columns=["ID", "Name", "Salary"])

print(df1)
```

```python
# This code iterates through the rows of a DataFrame using iterrows() and prints the index and values of each row.
data = {
    "Name": ["John", "Sam", "Alice"],
    "Age": [25, 30, 28]
}

df = pd.DataFrame(data)

# using iterrows()

for index, row in df.iterrows():
    print(index, row["Name"], row["Age"])
```

```python
# This code filters the DataFrame to include only rows where the "Age" column is greater than 25 and prints the result.
result = df[df["Age"] > 25]

print(result)
```

```python
# This code accesses the second row of the DataFrame using iloc and prints it.
print(df.iloc[1])     # second row
```

```python
# This code accesses the "Name" column for the first two rows of the DataFrame using loc and prints it.
print(df.loc[0:1, ["Name"]])
```

```python
# This code filters the DataFrame to include only rows where the "Age" column is greater than or equal to 28 and prints the result.
result = df[df["Age"] >= 28]

print(result)
```

```python
# This code adds a new row to the DataFrame at index 1 and prints the updated DataFrame.
data = {
    "Name": ["John", "Sam"],
    "Age": [25, 30]
}

df = pd.DataFrame(data)

new_row = pd.DataFrame({
    "Name": ["Alice"],
    "Age": [28]
})

df = pd.concat([df.iloc[:1], new_row, df.iloc[1:]]).reset_index(drop=True)

print(df)
```
