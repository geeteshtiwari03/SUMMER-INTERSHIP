```python
# 1) Convert a series of date-strings to a timeseries

import pandas as pd

dates = ["2026-01-01", "2026-01-02", "2026-01-03", "2026-01-04"]

ts = pd.to_datetime(dates)

print(ts)
```

```python
# 2) Create two DataFrames and perform merge operations

import pandas as pd

df1 = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["Aman", "Rohit", "Karan", "Vikas"]
})

df2 = pd.DataFrame({
    "ID": [2, 3, 4, 5],
    "Marks": [80, 90, 85, 75]
})

print("DataFrame 1")
print(df1)

print("\nDataFrame 2")
print(df2)

# Inner Merge
inner_merge = pd.merge(df1, df2, on="ID", how="inner")

print("\nInner Merge")
print(inner_merge)
```

```python
# Left Join

left_join = pd.merge(df1, df2, on="ID", how="left")

print(left_join)
```

```python
# Right Join

right_join = pd.merge(df1, df2, on="ID", how="right")

print(right_join)
```

```python
# Index-Based Join using join()

df1_index = df1.set_index("ID")
df2_index = df2.set_index("ID")

index_join = df1_index.join(df2_index, how="inner")

print(index_join)
```

```python
# Merging with Multiple Keys

df3 = pd.DataFrame({
    "ID": [1, 1, 2, 2],
    "Subject": ["Math", "Science", "Math", "Science"],
    "Marks": [80, 85, 90, 95]
})

df4 = pd.DataFrame({
    "ID": [1, 2, 2],
    "Subject": ["Math", "Math", "Science"],
    "Grade": ["A", "A+", "A"]
})

multi_merge = pd.merge(df3, df4, on=["ID", "Subject"])

print(multi_merge)
```

```python
# 3) Concatenate and Merge DataFrames

df5 = pd.DataFrame({
    "ID": [1, 2],
    "Name": ["A", "B"]
})

df6 = pd.DataFrame({
    "ID": [3, 4],
    "Name": ["C", "D"]
})

df7 = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "City": ["Delhi", "Mumbai", "Jaipur", "Pune"]
})

# Vertical Concatenation
concat_df = pd.concat([df5, df6], ignore_index=True)

print("Concatenated DataFrame")
print(concat_df)

# Merge with third DataFrame
merged_df = pd.merge(concat_df, df7, on="ID")

print("\nMerged DataFrame")
print(merged_df)
```

```python
# Difference Between join() and merge()

# merge()
# 1. Used for combining DataFrames using columns
# 2. More flexible
# 3. Supports inner, left, right, outer joins

# join()
# 1. Mainly used for joining based on index
# 2. Simpler syntax
# 3. Faster for index-based joins
```
