```python
# 1) Repeat a tuple three times using the * operator

t = (1, 2, 3)

result = t * 3

print("Repeated Tuple:", result)

```

```python
# 2) Join three separate tuples into one new tuple using the + operator

t1 = (1, 2)
t2 = (3, 4)
t3 = (5, 6)

result = t1 + t2 + t3

print("Joined Tuple:", result)
```

```python
# 3) Check whether a specific element exists inside a tuple using the in keyword

t = (10, 20, 30, 40)

element = 30

if element in t:
    print("Element exists in the tuple")
else:
    print("Element does not exist in the tuple")
```

```python
# Calculate total, highest value, and lowest value from a tuple without using sum(), max(), and min()

t = (12, 45, 7, 89, 23)

total = 0
highest = t[0]
lowest = t[0]

for num in t:
    total += num

    if num > highest:
        highest = num

    if num < lowest:
        lowest = num

print("Total =", total)
print("Highest =", highest)
print("Lowest =", lowest)
```

```python
# 5) Filter a tuple and keep only values greater than 10

n = (3, 14, 7, 22, 9, 41, 18, 5)

result = ()

for i in n:
    if i > 10:
        result += (i,)

print("Filtered Tuple:", result)
```

```python
 # 6) Determine how many elements are in a set without using len()

s = {"cat", "dog", "bird", "fish"}

count = 0

for item in s:
    count += 1

print("Number of elements:", count)
```

```python
# 7) Combine two sets into one containing all unique elements

s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

result = s1 | s2

print("Combined Set:", result)
```

```python
# 8) Find all elements that are common to both sets

s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

common = s1 & s2

print("Common Elements:", common)
```

```python
# 9) Find all elements that are in either set but not in both (Symmetric Difference)

s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

result = s1 ^ s2

print("Elements in either set but not both:", result)
```
