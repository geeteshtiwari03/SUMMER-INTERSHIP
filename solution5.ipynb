```python
import csv


file_name = "address_book.csv"


n = int(input("How many records you want to enter: "))


with open(file_name, mode='w', newline='') as file:
    writer = csv.writer(file)

    
    writer.writerow(["Name", "Address", "Mobile", "Email"])

    
    for i in range(n):
        print(f"\nEnter details for Person {i+1}")

        name = input("Name: ")
        address = input("Address: ")
        mobile = input("Mobile: ")
        email = input("Email: ")

        writer.writerow([name, address, mobile, email])

print("\nData stored successfully in address_book.csv")


print("\nStored Records:\n")

with open(file_name, mode='r') as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```

```python
import sqlite3

conn = sqlite3.connect("college.db")

cur = conn.cursor()


cur.execute("""
CREATE TABLE IF NOT EXISTS student(
    id INTEGER PRIMARY KEY,
    name TEXT,
    city TEXT
)
""")

cur.execute("""
CREATE TABLE IF NOT EXISTS course(
    cid INTEGER PRIMARY KEY,
    cname TEXT,
    fees INTEGER
)
""")

cur.execute("""
CREATE TABLE IF NOT EXISTS employee(
    eid INTEGER PRIMARY KEY,
    ename TEXT,
    salary INTEGER
)
""")

print("Tables created successfully")

cur.execute("INSERT INTO student VALUES(1,'John','Delhi')")
cur.execute("INSERT INTO student VALUES(2,'Rahul','Mumbai')")

cur.execute("INSERT INTO course VALUES(101,'Python',5000)")
cur.execute("INSERT INTO course VALUES(102,'Java',6000)")

cur.execute("INSERT INTO employee VALUES(1,'Amit',25000)")
cur.execute("INSERT INTO employee VALUES(2,'Sara',30000)")

conn.commit()

print("Records inserted successfully")

print("\nStudent Table:")
res = cur.execute("SELECT * FROM student")

for row in res:
    print(row)

print("\nCourse Table:")
res = cur.execute("SELECT * FROM course")

for row in res:
    print(row)

print("\nEmployees with salary > 25000")
res = cur.execute("SELECT * FROM employee WHERE salary > 25000")

for row in res:
    print(row)

cur.execute("UPDATE employee SET salary = 35000 WHERE eid = 2")
conn.commit()

print("\nEmployee data updated")


cur.execute("DELETE FROM student WHERE id = 2")
conn.commit()

print("\nStudent record deleted")


print("\nFinal Student Table:")
res = cur.execute("SELECT * FROM student")

for row in res:
    print(row)

conn.close()

print("\nDatabase connection closed")
```
